Q3. What errors are defined in the ArithmeticError class? Explain any two with an example.

Ans -

The ArithmeticError class is a base class for exceptions that are related to arithmetic operations. Two common exceptions derived from the ArithmeticError class are ZeroDivisionError and OverflowError. 

- ZeroDivisionError

```python
try:
    2/0
except ZeroDivisionError as e:
    print(f"{e}, {e.__class__}")
```

- OverflowError

```python
import math
try:
    print(math.exp(1000))
except OverflowError as e:
    print(f"{e}, {e.__class__}")
```

Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.

Ans - 


The LookupError class in Python is a base class for exceptions that are raised when a key or index is not found during a lookup operation. It's a parent class for exceptions like KeyError and IndexError, which are used to handle specific lookup-related errors. 

- Key Error

```python
a = {"name" : "pranjal", "class" : "Bsc 2nd Year"}

try:
    b = a["age"]
except KeyError as e:
    print(f"{e}, {e.__class__}")
```

- Index Error

```python
s = [2, 3, 4, 5, 7]
try:
    print(s[9])
except IndexError as e:
    print(f"{e}, {e.__class__}")
```

Q5. Explain ImportError. What is ModuleNotFoundError?

Ans -

- ImportError is a general exception that occurs when an error is encountered during the import of a module or package. This exception can have various subtypes or specific error messages depending on the nature of the problem.
- ModuleNotFoundError is a specific subtype of ImportError introduced in Python 3.6.
It is raised when Python cannot locate the module or package you are trying to import.
This error message provides more specific information about which module couldn't be found.
